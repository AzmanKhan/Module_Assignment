```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("dataset.csv")
```

```python
# univariate analysis
plt.figure(figsize=(5,2))
sns.histplot(df['AGE'],kde =True , color = 'skyblue')
plt.title("Distribution of age")
plt.xlabel('AGE')
plt.ylabel('Frequency')
plt.show()
```

```python
df.describe()
```

```python
# how many smoker have lung cancer 
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='SMOKING',hue='LUNG_CANCER', palette='viridis')
plt.title('Lung Cancer Diagnosis by Smoker')
plt.xlabel('SMOKING')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.show()
```

```python

# Set the style for the plots
sns.set(style="whitegrid")


# Univariate Analysis
# Numerical Feature: Age
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
# Categorical Feature: Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='GENDER', palette='viridis')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```

```python
# Bivariate Analysis
# Gender vs Lung Cancer
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='GENDER', hue='LUNG_CANCER', palette='viridis')
plt.title('Lung Cancer Diagnosis by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.show()

```

```python
# Smoking vs Lung Cancer
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='SMOKING', hue='LUNG_CANCER', palette='coolwarm')
plt.title('Lung Cancer Diagnosis by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.show()

```

```python
# Age vs Lung Cancer
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='LUNG_CANCER', y='AGE', palette='magma')
plt.title('Age Distribution by Lung Cancer Diagnosis')
plt.xlabel('Lung Cancer Diagnosis')
plt.ylabel('Age')
plt.show()

```

```python
# Multivariate Analysis
# Correlation Matrix
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Feature Interactions
# Gender vs Smoking
print(pd.crosstab(df['GENDER'], df['SMOKING']))

# Gender vs Age by Lung Cancer
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='GENDER', y='AGE', hue='LUNG_CANCER', palette='Set2')
plt.title('Age Distribution by Gender and Lung Cancer Diagnosis')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Lung Cancer')
plt.show()

```

```python
# Missing Values
print(df.isnull().sum())
```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='LUNG_CANCER', y='AGE', palette='Set2')
plt.title('Age Distribution by Lung Cancer Diagnosis')
plt.xlabel('Lung Cancer Diagnosis')
plt.ylabel('Age')
plt.show()
```
